```python
import pandas as pd

# Create a simple air quality dataset
data = {
    'City': ['Lagos', 'Abuja', 'Kano', 'Port Harcourt', 'Ibadan', 'Enugu', 'Kaduna', 'Benin City'],
    'PM2.5': [55, 40, 65, 80, 50, 45, 60, 70],
    'PM10': [120, 90, 150, 200, 110, 95, 140, 180],
    'NO2': [35, 25, 40, 45, 30, 28, 38, 42],
    'SO2': [15, 10, 18, 20, 12, 11, 17, 19],
    'CO': [1.2, 0.9, 1.5, 1.8, 1.1, 1.0, 1.3, 1.6]
}

df = pd.DataFrame(data)

# Save it as a CSV file (optional, for later download)
df.to_csv('air_quality.csv', index=False)

df.head()
```

### 🧠 Problem Statement
Air pollution is a growing problem in many urban areas, especially in developing countries.  
This project aims to group Nigerian cities based on their air quality metrics to help policymakers  
identify which regions need the most urgent intervention.

---

### 🎯 Objective
Use **K-Means Clustering (Unsupervised Learning)** to:
- Analyze air quality data (PM2.5, PM10, NO2, SO2, CO)
- Identify clusters of cities with similar pollution patterns
- Support data-driven environmental policy planning

---


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Normalize data
features = df[['PM2.5', 'PM10', 'NO2', 'SO2', 'CO']]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize clusters
plt.figure(figsize=(8,6))
plt.scatter(df['PM2.5'], df['PM10'], c=df['Cluster'], cmap='viridis', s=100)
plt.title('Air Quality Clustering by Pollution Levels (SDG 11)')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.show()

df
```

## 📊 Results & Insights

The clustering model successfully grouped cities based on their pollution levels.

- **Cluster 0:** Cities with moderate pollution (e.g., Abuja, Enugu)
- **Cluster 1:** Highly polluted cities (e.g., Port Harcourt, Kano)
- **Cluster 2:** Relatively cleaner air (e.g., Ibadan)

These insights can guide sustainable urban planning, environmental policy decisions,  
and health awareness programs in Nigeria.
